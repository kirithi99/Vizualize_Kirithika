```python
df.head()
```

```python
import seaborn as sns
```

HISTOGRAM PLOT

```python
sns.histplot(df['Open'],kde=True,color='green')
plt.title('Price at which stock trading starts')
plt.gca().set_facecolor('lightblue')
```

PAIRPLOT

```python
sns.pairplot(data=df)
```

BAR PLOT

```python
sns.barplot(x='Low',y='High',data=df,color='red')
plt.title("Relation between Low Price and High Price of stock trading day")
```
